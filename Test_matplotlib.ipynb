# MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4])
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('funny numbers')
```

```python
plt.plot([1, 2, 3, 4],[2, 3, 4, 5])
plt.ylabel('angry numbers')
```

```python
plt.plot([1, 3, 4, 9],[-1, 8, 1, 2])
```

```python
plt.plot([1, 3, 4, 9],[-1, 8, 1, 2], 'bo')
plt.plot([1, 9, 3, 2],[-6, 2, 3, 1], 'r+')
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', t, t**2 ,'bs', t, t**3,'g^', t,t**4, 'r-.')
```

```python
data = {
    'a': np.arange(500),
    'b': np.random.randint(0, 50, 50),
    'c': np.random.rand(500)
}
```
